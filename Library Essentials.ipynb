##Numpy:

```python
import numpy as np
```

```python
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)
print(a)
print(b)
```

```python
a = a.reshape(5, 4)
b = b.reshape(5, 4)

print("a:\n", a)
print("b:\n", b)
```

```python
scaler = 2 #chosen randomly

a = a * scaler
b = b * scaler

print("a:\n", a)
print("b:\n", b)
```

```python
b = b.T #Transpose
c = np.dot(a, b)
print("c:\n", c)
```

```python
c_sliced = c[:, 1:3]
print("Columns 2 and 3 of c:\n",c_sliced,"\n")
print("Whole matrix c:\n", c[:])
```

```python
max_element = c.max()
print(max_element)

index = np.where(c == max_element)
print(index)
```

```python
one_dim_c = c.flatten()
print("One dimensional matrix c:\n",one_dim_c)
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

plt.figure(figsize=(10,8))
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*', s=80)
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity in Dhaka')
plt.grid(True)
plt.show()
```

```python
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks_1 = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12,8))
plt.plot(study_hours, marks_1 , marker='o', linestyle='-', color='magenta')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Maks')
plt.grid(True)
plt.show()
```

```python
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.figure(figsize=(10, 8))
plt.barh(subjects,marks)
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.title("Horizontal Bar Graph of Subject and Marks")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.bar(subjects,marks,color=['gray', 'blue', 'brown', 'purple', 'cyan'])
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Vertical Bar Graph of Subject Scores")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))


axes[0, 0].scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*',s=100)
axes[0, 0].set_xlabel('Temperature')
axes[0, 0].set_ylabel('Humidity')
axes[0, 0].set_title('Temperature vs Humidity in Dhaka')

axes[0, 1].plot(study_hours, marks_1 , marker='o', linestyle='-', color='magenta')
axes[0, 1].set_title("Study Hours vs Marks")
axes[0, 1].set_xlabel('Study Hours')
axes[0, 1].set_ylabel('Marks')

axes[1, 0].barh(subjects,marks)
axes[1, 0].set_title("Subject Scores (Horizontal)")
axes[1, 0].set_xlabel('Subjects')
axes[1, 0].set_ylabel('Marks')

axes[1, 1].bar(subjects,marks,color=['gray', 'blue', 'brown', 'purple', 'cyan'])
axes[1, 1].set_title("Subject Scores (Vertical)")
axes[1, 1].set_xlabel('Subjects')
axes[1, 1].set_ylabel('Marks')


plt.tight_layout()
plt.show()
```
